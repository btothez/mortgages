```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
from nltk.corpus import wordnet as wn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score
```

```python
np.random.seed(500)
```

```python
help(word_tokenize)
```

```python
df.dropna(inplace=True)
```

```python
train_x, test_x, train_y, test_y = model_selection.train_test_split(df[1], df[0],test_size=0.2)
```

```python
encoder = LabelEncoder()
enc_train_y = encoder.fit_transform(train_y)
enc_test_y = encoder.fit_transform(test_y)
```

```python
encoder.classes_
```

```python
enc_test_y
```

```python
help(TfidfVectorizer)
```

```python
tfidf_vect = TfidfVectorizer(ngram_range=(1,3))
tfidf_vect.fit(df[1])
```

```python
train_x_tfidf = tfidf_vect.transform(train_x)
test_x_tfidf = tfidf_vect.transform(test_x)
```

```python
# fit the training dataset on the NB classifier
nb = naive_bayes.MultinomialNB()
nb.fit(train_x_tfidf, train_y)

# predict the labels on validation dataset
predictions_nb = nb.predict(test_x_tfidf)

# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_nb, test_y)*100)
```
